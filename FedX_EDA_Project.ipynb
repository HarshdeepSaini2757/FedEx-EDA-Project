# **Project Summary -**

This project presents a comprehensive exploratory data analysis (EDA) of FedEx's Supply Chain Management System (SCMS) delivery history dataset, encompassing 4,592 global shipments across 28 countries. The analysis examines critical logistics metrics including delivery times, freight costs, shipment modes, and on-time performance to extract actionable insights for supply chain optimization.
Dataset Overview
The analysis is based on the SCMS Delivery History Dataset containing 10,324 original records with 33 columns. After rigorous data cleaning and feature engineering, the dataset was refined to 4,592 records with 42 columns, including 9 newly derived features. This comprehensive dataset covers multiple dimensions of global logistics operations, including shipment modes (Air, Ocean, Air Charter, Truck), destinations across 28 countries, various fulfillment channels, and vendor relationships.
Objectives and Scope
The primary objective of this analysis is to identify patterns, inefficiencies, and optimization opportunities within FedEx's global supply chain operations. Specific goals include: (1) analyzing delivery performance across different shipment modes and geographic regions, (2) identifying cost drivers and opportunities for cost reduction, (3) calculating efficiency metrics for benchmarking, (4) highlighting risk areas and underperforming routes, and (5) providing data-driven recommendations for operational improvement. The analysis addresses key business questions: What delivery time patterns exist across shipment modes? How do freight costs vary by mode, weight, and destination? What factors most impact on-time performance? Where are the primary cost optimization opportunities?
Methodology
The analysis follows a structured, professional EDA approach. The data wrangling phase involved converting date columns to datetime format, removing duplicate records, handling missing values appropriately, converting text-based numeric columns to proper numeric types, and validating for physically impossible values. Feature engineering created nine derived features: delivery_days, on_time_delivery indicator, cost_per_kg, total_shipment_cost, freight_pct_of_value, shipment_month, shipment_quarter, shipment_year, and day_of_week. The visualization phase implemented 22 strategic charts following the UBM (Univariate-Bivariate-Multivariate) framework to systematically explore data relationships and patterns.
Key Findings
Analysis revealed that delivery times range from 12 to 616 days with an average of 117.8 days. Freight costs average $11,397.18 per shipment with significant variation across modes. Air freight is the fastest option (median 95 days) but most expensive, while ocean freight is slowest (median 170 days) but most cost-effective. Geographic analysis identified Vietnam, South Africa, and Haiti as top-volume destinations. A moderate positive correlation (0.393) between weight and freight cost was identified, indicating weight-based pricing but also other cost factors. On-time delivery performance varies by mode, with different rates suggesting optimization opportunities.
Business Value and Impact
This analysis provides FedEx with critical data-driven insights for strategic decision-making. The findings enable identification of high-cost routes and underperforming suppliers for targeted negotiation. Shipment mode analysis supports cost-efficiency decisions, potentially reducing overall logistics costs by 5-10%. Performance benchmarking by country and supplier enables better vendor evaluation and relationship management. Temporal patterns reveal seasonal trends for improved resource allocation and capacity planning. The comprehensive cost analysis, particularly freight costs as percentage of product value, directly informs pricing strategy and profitability optimization.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Make a copy to preserve original
df_clean = df.copy()
print(f"\nOriginal dataset shape: {df.shape}")

# --- CONVERT DATE COLUMNS ---
print("\n--- Converting Date Columns ---")

date_columns = ['PQ First Sent to Client Date', 'PO Sent to Vendor Date',
                'Scheduled Delivery Date', 'Delivered to Client Date', 'Delivery Recorded Date']

for col in date_columns:
    if col in df_clean.columns:
        df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')
        print(f"✓ Converted {col}")

# --- REMOVE EXACT DUPLICATES ---
print("\n--- Removing Duplicate Rows ---")
initial_rows = len(df_clean)
df_clean = df_clean.drop_duplicates()
rows_removed = initial_rows - len(df_clean)
print(f"✓ Removed {rows_removed} duplicate rows")

# --- REMOVE ROWS WITH MISSING CRITICAL DATES ---
print("\n--- Handling Missing Critical Data ---")
initial_rows = len(df_clean)
df_clean = df_clean.dropna(subset=['Delivered to Client Date', 'PO Sent to Vendor Date'])
rows_removed_dates = initial_rows - len(df_clean)
print(f"✓ Removed {rows_removed_dates} rows with missing delivery dates")

# --- REMOVE IMPOSSIBLE VALUES ---
# --- CONVERT NUMERIC COLUMNS FROM STRING TO NUMBER ---
print("\n--- Converting String Columns to Numeric ---")

numeric_columns = ['Weight (Kilograms)', 'Line Item Quantity', 'Freight Cost (USD)',
                   'Line Item Insurance (USD)', 'Line Item Value']

for col in numeric_columns:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
        print(f"✓ Converted {col} to numeric")

# --- REMOVE ROWS WITH NaN AFTER CONVERSION ---
print("\n--- Removing Rows with Non-Numeric Values ---")
initial = len(df_clean)
df_clean = df_clean.dropna(subset=numeric_columns)
rows_removed_conversion = initial - len(df_clean)
print(f"✓ Removed {rows_removed_conversion} rows with non-numeric values")

print("\n--- Removing Physically Impossible Values ---")
initial = len(df_clean)
# Add these 2 lines to your existing code:
df_clean = df_clean[df_clean['Freight Cost (USD)'] >= 0]
print(f"✓ Removed rows with negative freight cost")

df_clean = df_clean[df_clean['Line Item Value'] > 0]
print(f"✓ Removed rows with invalid line item value")

# Weight must be positive
df_clean = df_clean[df_clean['Weight (Kilograms)'] > 0]
print(f"✓ Removed rows with weight ≤ 0")

# Quantity must be positive
df_clean = df_clean[df_clean['Line Item Quantity'] > 0]
print(f"✓ Removed rows with quantity ≤ 0")

total_removed = initial - len(df_clean)
print(f"Total rows removed for impossible values: {total_removed}")

# --- CREATE DERIVED FEATURES ---
print("\n--- Creating Derived Features ---")

# 1. Delivery time in days
df_clean['delivery_days'] = (df_clean['Delivered to Client Date'] -
                             df_clean['PO Sent to Vendor Date']).dt.days
print("✓ Created delivery_days (Delivered - PO Sent)")

# 2. On-time delivery indicator
df_clean['on_time_delivery'] = (df_clean['Delivered to Client Date'] <=
                                df_clean['Scheduled Delivery Date']).astype(int)
print("✓ Created on_time_delivery (0=late, 1=on-time)")

# 3. Cost per kilogram
df_clean['cost_per_kg'] = df_clean['Freight Cost (USD)'] / (df_clean['Weight (Kilograms)'] + 0.01)
print("✓ Created cost_per_kg")

# 4. Total shipment cost
df_clean['total_shipment_cost'] = df_clean['Freight Cost (USD)'] + df_clean['Line Item Insurance (USD)']
print("✓ Created total_shipment_cost")

# 5. Freight cost as percentage of product value
df_clean['freight_pct_of_value'] = (df_clean['Freight Cost (USD)'] /
                                     (df_clean['Line Item Value'] + 1)) * 100
print("✓ Created freight_pct_of_value")

# 6. Extract time components
df_clean['shipment_month'] = df_clean['PO Sent to Vendor Date'].dt.month
df_clean['shipment_quarter'] = df_clean['PO Sent to Vendor Date'].dt.quarter
df_clean['shipment_year'] = df_clean['PO Sent to Vendor Date'].dt.year
print("✓ Created shipment_month, shipment_quarter, shipment_year")

# 7. Day of week
df_clean['day_of_week'] = df_clean['PO Sent to Vendor Date'].dt.day_name()
print("✓ Created day_of_week")

# --- FINAL VALIDATION ---
print("\n--- Data Validation ---")
print(f"✓ All delivery times ≥ 0: {(df_clean['delivery_days'] >= 0).all()}")
print(f"✓ On-time values are 0/1: {set(df_clean['on_time_delivery'].unique()).issubset({0, 1})}")
print(f"✓ No negative costs: {(df_clean['Freight Cost (USD)'] >= 0).all()}")

print("\n" + "="*70)
print(f"Final cleaned dataset shape: {df_clean.shape}")
print(f"Total rows removed: {df.shape[0] - df_clean.shape[0]}")
print("="*70)

# Display new columns created
print(f"\nNew columns created:")
new_cols = [col for col in df_clean.columns if col not in df.columns]
for col in new_cols:
    print(f"  - {col}")
````
